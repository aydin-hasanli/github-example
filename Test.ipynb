```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library

df_survey = pd.read_csv('https://cocl.us/datascience_survey_data')
print ('Data read into a pandas dataframe!')
df_survey.rename(columns={'Unnamed: 0':''},inplace=True)
df_survey.head()
df_survey.set_index('', inplace=True)
```

```python
#df_survey.sort_values(by='Very interested', ascending=False, inplace=True)
#df_survey.head()
df_survey_per=(df_survey/22.33).round(2)
df_survey_per.head()
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
ax=df_survey_per.plot(kind='bar', figsize=(20, 8), width= 0.8, color=['#5cb85c','#5bc0de','#d9534f'],edgecolor=None)
ax = plt.gca()
ax.spines['right'].set_visible(False)

ax.spines['top'].set_visible(False)

ax.spines['left'].set_visible(False)
plt.tick_params(

axis='y', # changes apply to the x-axis

which='both', # both major and minor ticks are affected

bottom=False, # ticks along the bottom edge are off

top=False, # ticks along the top edge are off

labelbottom=False, # labels along the bottom edge are off

right=False,

left=False,

labelleft=False)


#use color #5cb85c for the Very interested bars, color #5bc0de for the Somewhat interested bars, and color #d9534f for the Not interested bars
#plt.xlabel('Year') # add to x-label to the plot
#.ylabel('Number of immigrants') # add y-label to the plot
plt.legend(fontsize='14')
plt.xticks(fontsize=14)
plt.title("Percentage of the Respondents' Interest in Data Science Area", fontsize='16') # add title to the plot
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.2%}'.format(height/100), (x, y + height + 1.00), fontsize='14')
plt.show()
```
